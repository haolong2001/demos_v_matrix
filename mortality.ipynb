```python
import numpy as np

# Assuming mortality_matrix is already defined
death_by_cohort = np.zeros((2,86, 34))  # male, female; 0 - 85, 85+; for 1990 - 2023

for i in range(2):
    death_by_year = mortality_matrix[i, :, :]
    # Expand every row 5 times to get mortality rate from 0 - 85+
    death_table_4 = np.repeat(death_by_year[1:2], 4, axis=0)
    full_death_table = np.repeat(death_by_year[2:-1], 5, axis=0)
    full_death_table = np.vstack((death_by_year[0],death_table_4, full_death_table, death_by_year[-1]))

    # Iterate through each row and generate the death_by_cohort
    for j in range(86):
    
        # modify and check boundary
        # initialize vec with 0 
        vec = np.zeros(34)
        # for the upper matrix 
        
        for k in range(34):
                row_idx = min(85, j+k)
                vec[k] = full_death_table[row_idx, k]
            # diagnosis first, and then horizontal
             
        death_by_cohort[i, j, :] = vec
        # Fill the rest of the row with the last available value
        
```

```python
import pandas as pd

# Create a list of years for the columns
years = list(range(1990, 2025))  # 1990 to 2024

# Create a list of age group labels for the rows
age_groups = [str(i) for i in range(85)] + ['85+']  # Age groups 0-85 and '85+'

# Convert the numpy array to a DataFrame, set column names to years and row names to age groups
comparison_df = pd.DataFrame(comparison_matrix, columns=years, index=age_groups)

# Save the DataFrame to a CSV file
comparison_df.to_csv('comparison_matrix.csv', index=True)

# Display a success message
print("The comparison matrix has been saved to 'comparison_matrix.csv' with the appropriate row and column names.")
```
