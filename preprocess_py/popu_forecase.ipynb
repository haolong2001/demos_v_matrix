```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

mortality_matrix = death_matrix[0,:, 20:28]
# Step 2: Fit the Lee-Carter Model
log_mortality = np.log(mortality_matrix)
a_x = np.mean(log_mortality, axis=1)
log_mortality_centered = log_mortality - a_x[:, None]
U, S, Vt = np.linalg.svd(log_mortality_centered, full_matrices=False)
b_x = U[:, 0]
k_t = S[0] * Vt[0]

# Step 3: Forecast k_t using ARIMA
model = ARIMA(k_t, order=(1, 1, 0))  # Simple ARIMA(1,1,0) model
model_fit = model.fit()
k_t_forecast = model_fit.forecast(steps=25)

# Step 4: Reconstruct future mortality rates
future_years = np.arange(2021, 2021 + len(k_t_forecast))
forecasted_mortality = np.exp(a_x[:, None] + b_x[:, None] * k_t_forecast)

# Step 5: Visualization

```
